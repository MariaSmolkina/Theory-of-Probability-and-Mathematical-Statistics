**Задача 1**

Даны значения зарплат из выборки выпускников: 
```
100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 230, 24, 57, 55, 70, 75, 65, 84, 90, 150
```
Используя только встроенные питоновские функции и структуры данных (т.е. без библиотек `numpy`, `pandas` и др.), 
посчитайте (несмещённое) среднее квадратичное отклонение для данной выборки.

````python
x = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 230, 24, 57, 55, 70, 75, 65, 84, 90, 150]
````

*Выборочное среднее:*

````python
def mean(sample):
  return sum(sample) / len(sample)
````

````python
mean(x)
````

*Cреднее квадратичное отклонение*

````python
def std(sample):
  return (sum([(i - mean(sample)) ** 2 for i in sample]) / (len(sample) - 1) ) ** 0.5
````

````python
std(x)
````

*Проверка*

````python
import numpy as np
````

````python
np.mean(x)
````

````python
np.std(x, ddof=1)
````

**Задача 2**

Найдите число выбросов в выборке из задачи 1. Для определения выбросов используйте методику как при построении "усов" в `boxplot`, однако, как и в задаче 1, пользоваться можно только встроенными функциями и структурами данных.

````python
def quantile(sample, order):

  n = len(sample)
  sample = sorted(sample)

  n_left_part = int(n * order)
  n_right_part = int(n * (1 - order))

  if n_left_part + n_right_part == n:
    return (sample[n_left_part] + sample[n_right_part - 1]) / 2
  elif n_left_part + n_right_part == (n - 1):
    return sample[n_left_part]
````

````python
q1, q3 = quantile(x, 0.25), quantile(x, 0.75)
q1, q3
````

````python
iqr = q3 - q1
iqr
````

````python
boxplot_range = (q1 - 1.5 * iqr, q3 + 1.5 * iqr)
boxplot_range
````

````python
outliers = [x[i] for i in range(len(x)) if (x[i] < boxplot_range[0]) | (x[i] > boxplot_range[1])]
outliers
````

Выбросы в задаче 1: 230, 150.

**Задача 3**

В университет на факультет A поступило 100 человек, на факультет B - 90 человек и на факультет C - 45 человек. Вероятность того, что студент с факультета A сдаст первую сессию, равна 0.6. Для студента с факультета B эта вероятность равна 0.7, а для студента с факультета C - 0.8. Случайным образом выбирается студент, сдавший первую сессию. Какое событие наиболее вероятно:
1. студент учится на факультете A,
2. студент учится на факультете B,
3. студент учится на факультете C?

Пусть X - сдал первую сессию. Тогда:

P(X|A) = 0.6

P(X|B) = 0.7

P(X|C) = 0.8

P(A) = 100 / (100 + 90 + 45)

P(B) = 90 / (100 + 90 + 45)

P(C) = 45/ (100 + 90 + 45)

1) P(A|X), 2) P(B|X), 3) P(C|X) = ?

По формуле: P(A|X) = P(X|A) * P(A) / P(X)

P(X) = P(X|A) * P(A) + P(X|B) * P(B) + P(X|C) * P(C)

````python
p_x_a = 0.6
p_x_b = 0.7
p_x_c = 0.8
````

````python
p_a = 100 / (100 + 90 + 45)
p_b = 90 / (100 + 90 + 45)
p_c = 45/ (100 + 90 + 45)
p_a, p_b, p_c
````

````python
p_x = p_x_a * p_a + p_x_b * p_b + p_x_c * p_c
p_x
````

````python
p_a_x = p_x_a * p_a / p_x
p_b_x = p_x_b * p_b / p_x
p_c_x = p_x_c * p_c / p_x
p_a_x, p_b_x, p_c_x
````

Наиболее вероятно, что студент учится на факультете В.
