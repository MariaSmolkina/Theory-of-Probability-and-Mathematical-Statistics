**Задача 1**

Имеется колода в 52 карты. Найти число возможностей вытянуть из неё 4 карты так, чтобы среди них был хотя бы один туз.

В условии: хотя бы один туз, следовательно варианты: 1 туз, 2 туза, 3 туза, 4 туза.

```python
number_poss_1 = scipy.special.comb(4, 1)*scipy.special.comb(48, 3)
number_poss_2 = scipy.special.comb(4, 2)*scipy.special.comb(48, 2)
number_poss_3 = scipy.special.comb(4, 3)*scipy.special.comb(48, 1)
number_poss_4 = scipy.special.comb(4, 4)*scipy.special.comb(48, 0)
number_poss = number_poss_1 + number_poss_2 + number_poss_3 + number_poss_4
number_poss
```

*Какова вероятность, что все выбранные вопросы знакомы студенту?*

**Задача 4**

Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.

Вопрос: какова вероятность, что вы действительно больны ей?

Подсказка: используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

Пусть А - заболел, B - тест положительный.
Тогда P(A|B) = P(B|A)*P(A)/P(B).

P(B) = P(B|A)*P(A) + P(B|Ā)*P(Ā).

```python
p_b = 0.99*0.001 + 0.01*0.999
p_b
```

```python
p = 0.99*0.001/p_b
p
```
