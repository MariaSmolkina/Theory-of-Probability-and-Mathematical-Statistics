**Задача 4**

Рост взрослого населения города X имеет нормальное распределение.

Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

а). больше 182 см

б). больше 190 см

в). от 166 см до 190 см

г). от 166 см до 182 см

д). от 158 см до 190 см

е). не выше 150 см или не ниже 190 см

ё). не выше 150 см или не ниже 198 см

ж). ниже 166 см.



a) P(X > 180) = F0(+ беск) - F0((180 - мю) / сигма)

F0 = (1/ sqr(2 pi)) * интеграл от 0 до x от (exp(-t^2 / 2))

```python
from scipy import integrate
import math
```

```python
mu = 174
sigma = 8
```

```python
def f(x):
  return math.exp(- (x ** 2) / 2)
```

```python
a = 0
b = (180 - mu) / sigma
```

```python
v, err = integrate.quad(f, a, b)
v
```

Вероятность того, что случайным образом выбранный взрослый человек имеет рост больше 182 см:

```python
P = 1/2 - (1/math.sqrt(2 * math.pi)) * v
P
```

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост больше 190 см:

```python
a = 0
b = (190 - mu) / sigma
v, err = integrate.quad(f, a, b)
P = 1/2 - (1/math.sqrt(2 * math.pi)) * v
P
```

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 190 см:

```python
a = (166 - mu) / sigma
b = (190 - mu) / sigma
v2, err = integrate.quad(f, 0, b)
v1, err = integrate.quad(f, 0, a)
P = (1/math.sqrt(2 * math.pi)) *v2 - (1/math.sqrt(2 * math.pi)) *v1
P
```

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 182 см:

```python
a = (166 - mu) / sigma
b = (182 - mu) / sigma
v2, err = integrate.quad(f, 0, b)
v1, err = integrate.quad(f, 0, a)
P = (1/math.sqrt(2 * math.pi)) *v2 - (1/math.sqrt(2 * math.pi)) *v1
P
```

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост не выше 150 см или не ниже 190 см:

не выше 150 см:

```python
b = (150 - mu) / sigma
v, err = integrate.quad(f, 0, b)
P_1 = (1/math.sqrt(2 * math.pi)) * v + 1/2
P_1
```

не ниже 190 см:

```python
a = 0
b = (190 - mu) / sigma
v, err = integrate.quad(f, a, b)
P_2 = 1/2 - (1/math.sqrt(2 * math.pi)) * v
P_2
```

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост не выше 150 см или не ниже 190 см:

P_1 + P_2

сумма площадей под графиками, так как или


```python
P = P_1 + P_2
P
```

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост не выше 150 см или не ниже 198 см:
